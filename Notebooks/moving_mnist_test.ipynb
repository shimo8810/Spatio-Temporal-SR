```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt

from chainer.dataset import dataset_mixin
%matplotlib inline
```

```python
data = np.load('../datasets/mnist_test_seq.npy')
print('length of sequence:{}, data size:{}, width:{}, height:{}'.format(*data.shape))
print('data type:', data.dtype)
```

```python
seqs = data.transpose(1, 0, 2, 3)
print(data.shape)
print(seqs.shape)
test_seq, train_seq = np.split(seqs, [10000//10])
np.save('../datasets/moving_mnist_train.npy', train_seq)
np.save('../datasets/moving_mnist_test.npy', test_seq)
```

```python
idx = 35
sidx = idx // 18
fidx = idx % 18
print('sequence:', sidx, ', frame:', fidx)
s = test_seq[sidx, fidx:fidx+3]
plt.imshow(s[0])
plt.show()
plt.imshow(s[1])
plt.show()
plt.imshow(s[2])
plt.show()
```
