```python
from pathlib import Path
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
mmnist_path = Path('../results/SeqVAE//mmnist/')
print(mmnist_path)
```

```python
logs = {}
plt.figure(figsize=(18,6))
ax1 = plt.subplot(1, 2, 1)
ax2 = plt.subplot(1, 2, 2)
for p in sorted(mmnist_path.glob('SeqVAE_latent100_ch4*')):
    log_path = p.joinpath('log')
    log_df = pd.read_json(log_path)
    ax1.plot(log_df['main/rec_loss'], label=p.name)
    ax2.plot(log_df['main/kl_loss'], label=p.name)

ax1.set_ylim(100, 400)
ax1.set_xscale('log')
ax1.set_yscale('log')
ax1.legend()
ax2.legend()
ax1.grid(which='both')
ax2.grid()
plt.show()
```
