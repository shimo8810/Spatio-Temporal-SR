```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
image = Image.open('./obj4__0.png')
image
```

```python
np.linspace(0.4, 2.0, 72)
```

```python
for i, x in enumerate(range(-64, 64, 8)):
    data = (1, 0, x, 0, 1, 0)
    img_tf = image.transform(image.size, Image.AFFINE, data, Image.BICUBIC)
    img_tf.save('f_{}.jpg'.format(i))
```

```python
for i, r in enumerate(np.arange(0.5, 1.5, 0.05)):
    m1, m2, _ = np.linalg.inv(np.array([[r, 0, 64-64*r], [0, r, 64-64*r],[0,0,1]]))
    data = np.hstack((m1,m2))
    img_tf = image.transform(image.size, Image.AFFINE, data, Image.BICUBIC)
    img_tf.save('bic_{}.jpg'.format(i))
```

```python
imgs = []
for i, r in enumerate(np.arange(0.5, 1.51, 0.5)):
    m1, m2, _ = np.linalg.inv(np.array([[r, 0, 64-64*r], [0, r, 64-64*r],[0,0,1]]))
    data = np.hstack((m1,m2))
    img_tf = image.transform(image.size, Image.AFFINE, data, Image.BICUBIC)
    imgs.append(np.array(img_tf))
```

```python
imgs = np.array(imgs)
```

```python
a = np.rot90(imgs, axes=(1,2))
plt.imshow(a[0])

plt.imshow(np.flip(imgs, 1)[0])
```
